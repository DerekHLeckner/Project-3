```python
# Load library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# Create URL to Excel file (alternatively this can be a filepath)
url = 'http://ir.eia.gov/ngs/ngshistory.xls'
# Load the second sheet of the Excel file into a data frame
natgasstorEIA_df = pd.read_excel(url, 'weekly_net_changes')
# Remove unwanted information
new_header = natgasstorEIA_df.iloc[5] #grab the first row for the header
natgasstorEIA_df = natgasstorEIA_df[6:] #take the data less the header row
natgasstorEIA_df.columns = new_header #set the header row as the df header
natgasstorEIA_df = natgasstorEIA_df.drop(['East Region','Midwest Region','Mountain Region','Pacific Region','South Central Region','Salt','NonSalt'], 1)
natgasstorEIA_df = natgasstorEIA_df.iloc[405:]
natgasstorEIA_df = natgasstorEIA_df.iloc[::-1]
natgasstorEIA_df
```

```python
z = pd.DataFrame(natgasstorEIA_df.dropna(axis=1, how='all'))
```

```python
# z = df.rename(columns={'Week ending': 'date'})
t = z.rename({'Week ending': 'date'}, axis=1)  # new method
```

```python
# Export file as a CSV 
t.to_csv("ye.csv", index=False, header=True)
```
