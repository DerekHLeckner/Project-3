```python
# Load library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import quandl
import pandas as pd, datetime as dt
import datetime
from datetime import datetime
from dateutil import parser
```

```python
# Read the csv & Change date format
df = pd.read_csv('historic_storage.csv', parse_dates = ['date'])
df
```

```python
# Using group by for how many location on csv file
df_loc = df.groupby('loc')
```

```python
# Print unique loc
print(df_loc.groups.keys())
```

```python
# We can look at the location individually

# FreebirdGasStorageL.L.C. = df_loc.get_group('0002')[['date','tsp_name','tsq']]
# FREE3 = df_loc.get_group('021039')[['date','tsp_name','tsq']]

# CaledoniaEnergyPartners = df_loc.get_group('542524')[['date','tsp_name','tsq']]

# BayGasStorage = df_loc.get_group('BG-1003')[['date','tsp_name','tsq']]
# B3 = df_loc.get_group('BG-1008')[['date','tsp_name','tsq']]
# B4 = df_loc.get_group('BG-1013')[['date','tsp_name','tsq']]

# We don't need to all location that's why we are getting one company and location
C = (pd.DataFrame(df_loc.get_group('542519')[['date','tsq']])).rename(columns={'tsq':'Caledonia Energy Partners, L.L.C.'})
B1 = (df_loc.get_group('BG-1002')[['date','tsq']]).rename(columns={'tsq':'BayGas Storage'})
EAST1 = (df_loc.get_group('029734466')[['date','tsq']]).rename(columns={'tsq':'East Cheyenne Gas Storage'})
FREE1 = (df_loc.get_group('012709')[['date','tsq']]).rename(columns={'tsq':'Freebird Gas Storage L.L.C.'})
```

```python
# Start to merge dataframe one by one
A=pd.merge(C, B1, on=["date"])
A.head()
```

```python
B=pd.merge(A,EAST1, on=["date"])
B.head()
```

```python
Historic=pd.merge(B,FREE1, on=["date"])
Historic
```

```python
# Export file as a CSV 
Historic.to_csv("Historic.csv", index=False, header=True)
```
